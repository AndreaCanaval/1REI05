### 1. Función de demanda de consumo


La función de consumo representa que todas las familias del país se ven afectadas por una forma doble de consumo: consumo autónomo (que es fijo, ya que la gente siempre consumirá una canasta básica) y un consumo variable (propensión marginal a consumir que varía de acuerdo al ingreso disponible).

$$ T=tY $$ 

Donde "t" es la tasa de impuesto o de tributación promedio.

$$ Y^d=Y-T $$

$$ Y^d=Y-T$$ 

Sabiendo que $T= tY$,entonces

$$ Y^d= (1-t)Y $$

Por lo tanto, donde una parte del ingreso disponible se utiliza para los gasto en consumo 

$$ C=C_0 + bY^d$$


En donde: 
    
Yd: ingreso disponible   

b: proporción marginal a consumir (PMgC) que esta entre 0 a 1.

Co: consumo autónomo (fijo)


$$ \frac{∆C}{∆Y^d} = b , donde  ( 0<b<1) $$


Por ello, de manera intuitiva se observa que:

$$Co↑ → DA↑ → DA > Y → Y↑$$

### 2. Función de Inversión

La demanda de Inversión, se refiere a gastos que aumenta el stock de capital y, por lo tanto, la producción potencial o
producción de largo plazo de la economía.
Si consideramos que el diferencial de la función de la Demanda de inversión:

$$ ∆I=∆I_0-∆hr$$

Sabiendo que los componentes autónomos se mantienen o no cambian. Entonces,tenemos que la demanda de inversión:

$$ I=I_0-hr$$


En donde:

Io : inversión autónoma 

h : sensibilidad de los inversionistas ante la subida o bajada de interés

r: tasa de interés real pasiva

### 2.1. Gráfico de la Función de Inversión

### 4. El Ingreso de equilibrio a corto plazo

### La Determinación del Ingreso de Equilibrio


Para la determinar el ingreso de equilibrio,sabemos que la demanda agregada se denota como:

$$ DA=C+I+G+X-M $$

Sabiendo que por dato:

$ C= C_0+bY^d$

$ I= I_0-hr$

$ G= G_0d$

$ T= tY$

$ X= X_0$

$ M= mY^d$


Por lo tanto:

$$ DA=(C_0+I_0+G_0+X_0-hr) + (b-m)(1-t)Y$$

En forma breve:

$$DA= \alpha_0 + \alpha_1Y$$

Donde $\alpha_0=(C_0+I_0+G_0+X_0-hr)$ es el intercepto y $\alpha_1=(b-m)(1-t)$ es la pendiente de la función.

Por ende, en equilibrio, el ingreso agregado,$Y$, debe ser igual a la demanda agregada $DA$. De la condición de equilibrio $Y=DA$, se obtiene la ecuación de ingreso o producto de equilibrio a corto plazo.

$$ Y=(C_0+I_0+G_0+X_0-hr) + (b-m)(1-t)Y$$    
$$ [1-(b-m)(1-t)]Y=(C_0+I_0+G_0+X_0-hr) $$
$$ Y= \frac{1}{1-(b-m)(1-t)}(C_0+I_0+G_0+X_0-hr)$$

A corto plazo se supone que:

- Los precios son fijos y que la economía se encuentra por debajo de la producción de pleno empleo.  
- La oferta agregada es infinitamente elástica al nivel de precios dado. 
- El producto e ingreso de equilibrio está determinado por la demanda agregada o por sus componentes. 
- El ajuste en el mercado de bienes es por cantidades y no por precios.



#### 4.2. Gráfico

```python
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

# Recta de 45°

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)

# Gráfico

# Dimensiones del gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(15, 8))

# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas punto de equilibrio
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.5, linestyle = ":", color = "grey")
plt.axhline(y=145, xmin= 0, xmax= 0.65, linestyle = ":", color = "grey")
# Texto agregado
    # punto de equilibrio
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')
    # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

# Título y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA',xlim=(0, 150))
ax.legend() #mostrar leyenda

plt.text(100,205, r'$DA= \alpha_0 + \alpha_1 Y $', fontsize = 15, color = 'black')
plt.text(98,190, '$←$', fontsize = 24, color = 'black',rotation=60)
plt.text(102,185, r'$\alpha_1 = (b-m)(1-t)$', fontsize = 15, color = 'black')
plt.text(75,115, r'$\alpha_0=(C_0 + I_0 + G_0 + X_0 -hr )  $', fontsize = 15, color = 'black')
plt.text(76,130, '$←$', fontsize = 24, color = 'black',rotation=240)
plt.show()
```

#### 5.1 Política fiscal expansiva con aumento del Gasto del Gobierno

- De manera intutiva:

Sabemos que, si el gobierno aumenta su gasto $∆G>0$, la Demanda Agregada tambien aumentaría, por lo que el Ingreso de equilibrio también aumenta

$$ ↑Go → ↑DA → DA =Y → ↑Y $$

- Matematicamente:
Si $$ ∆Go>0 →  ¿∆Y? $$

Sabiendo que 

$$ Y_e = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

Si: $ k=\frac{1}{1 - (b - m)(1 - t)} $ , en donde $ k > 0 $ , tenemos lo siguiente:

$$ Y = k (C_0 + I_0 + G_0 + X_0 - hr) $$
$$ ∆Y = k (∆C_0 + ∆I_0 + ∆G_0 + ∆X_0 - ∆hr) $$

Como solo aumentó el gasto, quiere decir que todos los demas son constantes, entonces 

$$∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0$$
$$ ∆Y = k (∆G_0) $$

Sabiendo que $∆G_0 > 0 $ y $k > 0$ ,entonces

$$k∆G_0 > 0$$
Reemplazando 

$$∆Y>0$$


```python
# Curva de ingreso de equilibrio ORIGINAL

    # Parámetros
    
Y_size = 100 

Co = 35
Io = 40
Go = 35
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9
Y = np.arange(Y_size)

    # Ecuación 
    
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

#--------------------------------------------------
# NUEVA curva de ingreso de equilibrio

    # Definir SOLO el parámetro cambiado
Go = 70

# Generar la ecuación con el nuevo parámetro
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))


# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #curva ORIGINAL
ax.plot(Y, DA_G, label = "DA_G", color = "#EE7600", linestyle = 'dashed') #NUEVA curva
ax.plot(Y, L_45, color = "#404040") #línea de 45º

# Lineas punteadas
plt.axvline(x = 70.5, ymin= 0, ymax = 0.69, linestyle = ":", color = "grey")
plt.axhline(y = 176, xmin= 0, xmax = 0.7, linestyle = ":", color = "grey")
plt.axvline(x = 54,  ymin= 0, ymax = 0.54, linestyle = ":", color = "grey")
plt.axhline(y = 135, xmin= 0, xmax = 0.54, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 180, '$DA_G$', fontsize = 11.5, color = '#EE7600')
plt.text(0, 125, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(72, 0, '$Y_G$', fontsize = 12, color = '#EE7600')
plt.text(56, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(60, 45, '$←$', fontsize = 23, color = 'grey',rotation=180)
plt.text(20, 140, '$↓$', fontsize = 23, color = 'grey',rotation=180)

# Título y leyenda
ax.set(title = "Aumento del Gasto del Gobierno $(G_0)$", xlabel = r'Y', ylabel = r'DA')
ax.legend()

plt.show()
```

#### Política fiscal expansiva con reducción de la Tasa de tributación


- De manera intutiva:¿contradicción?

$$ t↓ → Co↑ → DA↑ → DA > Y → Y↑ $$
$$ t↓ → M↑ → DA↓ → DA < Y → Y↓  $$

Con respecto a los dos escenarios (uno de consumo y otro de exportaciones e importaciones), se plantea que puede haber una contradicción entre ellos. Esto debido a que, con respecto al primer escenario, se observa un incremento en el nivel de producción debido a la reducción de la tasa impositiva. Asimismo, el grado de consumo por parte de la población puede incrementarse debido a que tendrán mas dinero al pagar menos impuestos. Esto quiere decir que tanto la demanda agregada como la de consumo aumentarían. A su vez, con respecto al segundo escenario, se observa que también puede darse el caso de que si la tasa de tributación se reduce, la demanda agregada lo hará también. Esto debido a la disminución de exportaciones.

- Matematicamente

Matemáticamente: $∆t < 0  →  ¿∆Y?$

$$ Y_e = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

Considernado el diferencial de $∆t$

$$ \frac{∆Y}{∆t} = \frac{(m-b)(Co + Go + Io + Xo - hr)}{(1-(1-t)(b-m)+1)^2} $$

- Sabiendo que b > m, entonces $(m-b) < 0$
- Los componentes autónomos no cambian: $∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0$
- Cualquier número elevado al cuadrado será positivo: $ (1-(1-t)(b-m)+1)^2 > 0 $

Entonces

$$ \frac{∆Y}{∆t} = \frac{(-)}{(+)} $$

Dado que $∆t < 0$, la división de dos positivos da otro positivo:

$$ \frac{∆Y}{(-)} = \frac{(-)}{(+)} $$

$$ ∆Y = \frac{(-)(-)}{(+)} $$

$$ ∆Y > 0 $$

####   Equilibrio con regla contracíclica
  
Primero tenemos las ecuaciones del gasto:

$ C = C_0 + bY^d $

$ I = I_0 - hr $

$ G = G_0 - gY $

$ X = X_0 $

$ M = mY^d $
    
Pero ahora el gasto del gobierno tiene un nuevo componente "$-gY$"
Si sumamos todas estas, encontramos la función de Demanda Agregada:

$$ DA = C + I + G + X - M$$
 
Ahora tomamos la función de Ingreso disponible:
$$ Y^d = Y(1 - t) $$

y la reemplazamos en la función de la demanda agregada:

$$ DA = C_0 + I_0 + G_0 + X_0 - hr + Y((b - m)(1 - t)-g) $$

Esta función se puede escribir de la siguiente manera, con dos partes alpha:
  
$$ DA = α_0 + α_0Y $$

$α_0 $sería el intercepto y $α_1$ sería la pendiente

Recordemos que el equilibrio es $DA = Y$, por lo que podemos decir que:

$$ Y = C_0 + bY^d + I_0 -hr + G_0 - gY + X_0 - mY^d $$

$$ Y^e = \frac{1}{1 - (b - m)(1 - t) + g} (C_0 + I_0 + G_0 + X_0 - hr) $$
 
Lo que cambia es que tenemos a "$g$" en el denominador del multiplicador keynesiano. Esta sería la ecuación de equilibrio del modelo Ingreso-Gasto keynesiano.


#### 6. 3. Política Fiscal Expansiva con aumento del Gasto de Gobierno
Intuitivo: 
$$ ↑Go → ↑DA → DA > Y → ↑Y $$


Se evidencia que a medida que el Go aumenta, la DA lo hará también. De esta forma, la DA logrará ser mayor que el ingreso agregado. Por ello, el ingreso deberá aumentar para lograr un equilibrio. Teniendo en cuenta el parámetro "$g$", la pendiente de la curva se reducirá. Sin embargo, esto obligará al ingreso agregado a aumentar para lograr nuevmente un equilibrio.
 
Matemático: $∆G_0 > 0  →  ¿∆Y?$ Debemos identificar cómo va a variar el Ingreso para regresar al equilibrio si aumenta el Gasto autónomo.

$$ Y = \frac{1}{1 - (b - m)(1 - t) + g} (C_0 + I_0 + G_0 + X_0 - hr) $$,

Ahora que tenemos a "$g$" en la ecuación, su efecto es que aumenta la magnitud del denominador, por lo que el tamaño de la pendiente se reduce.

Como no ha habido cambios en $C_0$, $I_0$, $X_0$, $h$ ni $r$, entonces la variación de Y solo se dará en términos del gasto autónomo: 

$$∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0$$

$$ ∆Y = k (∆G_0) $$

Sabiendo que $∆G_0 > 0 $  y que $k > 0$, la multiplicación de un número positivo con otro positivo dará positivo:

$$ ∆Y = (+)(+) $$

$$ ∆Y > 0 $$

Es así como el ingreso se incrementará en respuesta al aumento del gasto del gobierno
