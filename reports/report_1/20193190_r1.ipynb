<mark>2. Funci√≥n de Inversi√≥n: </mark>

Para conseguir la funci√≥n de inversi√≥n, se debe considerar el gasto en inversi√≥n (I), el cual aumenta el stock de capital en la econom√≠a y, por ende, incrementa tambi√©n su producci√≥n a largo plazo.

Este gasto incluye la variaci√≥n de existencia de las empresas y la inversi√≥n bruta fija, que a su vez considera la reposici√≥n del capital gastado y la inversi√≥n neta. 

Finalmente, la funci√≥n de demanda de inversi√≥n, adem√°s de representar el gasto de inversi√≥n, depende inversamente de la tasa de inter√©s real (r ) y de la Inversi√≥n aut√≥noma (Io), que depende de las expectativas de los inversionistas (h); por lo que se expresa de la siguiente manera: 

                                                     I = Io ‚àí hr
                                                     
                                                     
De esta manera, se entiende que existe una relaci√≥n negativa entre la inversi√≥n en un pa√≠s y la tasa de inter√©s, por lo que, si aumenta la inversi√≥n aut√≥noma, aumentar√° tambi√©n la inversi√≥n en general del pa√≠s, y viceversa. 

```python
r_size = 100 

Io = 40
h = 0.7

r = np.arange(r_size)

def I(Io, r, h):
    I = Io-h*r
    return I

I1= I(Io, r, h)
I1
```

- Pol√≠tica Fiscal Expansiva con una reducci√≥n de la Tasa de Tributaci√≥n $( ‚ñ≥t<0)$. 

   * An√°lisis intuitivo: 
      
   $$t‚Üì ‚Üí  Co‚Üë ‚Üí DA‚Üë ‚Üí DA > Y ‚Üí Y‚Üë$$
   
   $$t‚Üì ‚Üí  M‚Üë ‚Üí DA‚Üì ‚Üí DA < Y ‚Üí Y‚Üì$$


Una reducci√≥n de la Tasa de Tributaci√≥n (t) llevar√° a un aumento en el nivel de consumo (Co) y, por ende, aumentar√° tambi√©n la demanda agregada (DA). De esta manera, la demanda agregada ser√° mayor al nivel de producci√≥n (Y) en el pa√≠s; para satisfacer dicha demanda y regresar al punto de equilibrio, el nivel de producci√≥n tambi√©n aumentar√°.

Sin embargo, la tasa de tributaci√≥n no solo afecta al nivel de consumo, sino que tambi√©n est√° vinculada con las importaciones y las exportaciones. As√≠ que, al disminuir, y como tiene una relaci√≥n directa con las importaciones (M), estas van a aumentar; pero las importaciones est√° inversamente relacionadas con la demanda agregada, as√≠ que la demanda caer√° y ser√° menor al nivel de producci√≥n. Entonces, se estar√≠a produciendo m√°s de lo que se est√° demandando. Para regresar a la situaci√≥n de equilibrio, el nivel de producci√≥n tendr√≠a que caer. 

Entonces, se nota que existe una contradicci√≥n, ya que los choques que produce la tasa de tributaci√≥n en el nivel de consumo, provoca que el nivel de producci√≥n aumente; pero por el lado de las exportaciones e importaciones, provocan que el nivel de producci√≥n disminuya. Por ello, se tendr√° que ver el an√°lisis matem√°tico para resolver esta contradicci√≥n. Aunque, tambi√©n se debe considerar que, en la econom√≠a de un pa√≠s, tiene mayor impacto el nivel de consumo que sus importaciones, as√≠ que se deber√≠a considerar el aumento en el nivel de producci√≥n dado por el aumento en el consumo.

* **Nivel de Ingreso de equilibrio ùëå·µâ:**

Para encontrar el nivel de ingreso de equilibrio, primero se debe considerar la ecuaci√≥n de la demanda agregada (DA): 

$$DA = C + I + G + X ‚àí M$$

La cual, al remplazar sus componentes, y considerando la regla contrac√≠clica, se expresa de la siguiente manera: 

$$ùê∑ùê¥ = (ùê∂o + ùêºo + ùê∫o + ùëão ‚àí ‚Ñéùëü) + [(ùëè ‚àí ùëö)(1 ‚àí ùë°)-g]ùëå$$

Considerando que, en el nivel de equilibrio, se cumple que la demanda agregada es igual al nivel de ingreso (DA = Y), la ecuaci√≥n quedar√≠a de la siguiente manera: 

                                     
                                     ùëå = (ùê∂o + ùêºo + ùê∫o + ùëão ‚àí ‚Ñéùëü) + [(ùëè ‚àí ùëö)(1 ‚àí ùë°)-g]ùëå
                                     
                                     
                                     {1 ‚àí [(ùëè ‚àí ùëö)(1 ‚àí ùë°)-g]}ùëå = ùê∂o + ùêºo + ùê∫o + ùëão ‚àí ‚Ñéùëü
                                     

Finalmente, la ecuaci√≥n que expresa el nivel de ingreso de equilibrio es la siguiente: 

                                     
                                     ùëå·µâ =          1            
                                         ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî (ùê∂o + ùêºo + ùê∫o + ùëão ‚àí ‚Ñéùëü)
                                          1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°)-g

* **Pol√≠tica Fiscal Expansiva con aumento del Gasto del Gobierno $( ‚ñ≥G>0)$.**
   
   * An√°lisis intuitivo: 
   
   $$G= Go - gY$$
   
   $$G‚Üë ‚Üí Go‚Üë ‚Üí g‚Üë ‚Üí DA‚Üë ‚Üí DA>Y ‚Üí Y‚Üë$$
   
En este caso, se nota un aumento del Gasto P√∫blico _(Go)_ cuando disminuye el nivel de producci√≥n _(-gY)_. Este aumento del gasto del gobierno, supone tambi√©n un incremento de la demanda agregada _(DA)_ del pa√≠s. Ello lleva a la econom√≠a a un nivel de desequilibrio, dado que la demanda agregada es mayor al nivel de producci√≥n. Por tanto, y para volver al punto de equilibrio, el aumento del gasto p√∫blico supone tambi√©n un est√≠mulo para el aumento del nivel de producci√≥n. Todo ello ocurre a pesar de que _"g"_ disminuye el impacto del incremento del gasto p√∫blico, es decir, la demanda agregada y el nivel de producci√≥n del pa√≠s aumentar√°n, pero en menor medida debido a dicho par√°metro. 

   
   * An√°lisis matem√°tico: 

```python
Co, Io, Go, Xo, h, r, b, m, t, g = symbols('Co Io Go Xo h r b m t g')

f = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t)-g)


df_Go = diff(f, Go)
df_Go #‚àÜY/‚àÜGo
```

Considerando el diferencial de ‚ñ≥Go: 

                                            ‚ñ≥Y            1
                                            ‚Äî‚Äî‚Äî = ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî

Los componentes aut√≥nomos no cambian:

                                            ‚ñ≥Co= ‚ñ≥Io= ‚ñ≥Xo= ‚ñ≥h= ‚ñ≥t= ‚ñ≥r= 0
                                            

Dado que ‚ñ≥Go < 0, la divisi√≥n de dos negativos da positivo:

                                            ‚ñ≥Y    (+)
                                            ‚Äî‚Äî‚Äî = ‚Äî‚Äî‚Äî‚Äî
                                            (-)    (-)

                                            ‚ñ≥Y    (+)(-)
                                                = ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
                                                    (-)

                                            ‚ñ≥Y > 0

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3 
r = 0.9
g = 0.2

Y = np.arange(Y_size)

     
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + (((b - m)*(1 - t)-g)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y)

#----------------------------------------------------

Go = 90

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + (((b - m)*(1 - t)-g)*Y)
    return DA_K

DA_Go = DA_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y)
```
