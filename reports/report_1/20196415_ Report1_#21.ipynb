1.Derive y explique la función de demanda de Consumo. Replique el gráfico siguiente:
    
- C=Gasto del consumo de las familias
- Co=consumo autonomo--> es tu consumo fijo
- b=propensión marginal a consumir 
- Yd= Ingreso disponible en la economía--> 
- Yd=Y-T--> El ingreso agrgado menos la tributación es el ingreso disponible
- T=tY--> Tributación total del país
 $$ \\ Yd=Y-tY=(1-t)Y \\ $$

 Entonces:

$$ \\  C= Co+(b)Yd= Co+b(1-t)Y\\ $$


```python
Y_size = 100 

Co = 35
b = 0.5
Yd = np.arange(Y_size)

def DA_K(Co, b, Y):
    DA_K = (Co+b*Yd)
    return DA_K

DA_K = DA_K(Co, b, Yd)
```

4.A partir de las siguientes ecuaciones
$$ \\DA=Y\\ $$

$$ \\DA=C+I+G+X-M\\ $$

$$ \\DA=Co+Io+Go+Xo-hr+Y(b-m)(1-t)\\ $$

$$ \\DA=α0+α1 Y\\ $$

$$ \\Co+Io+Go+Xo-h*r+(b-m)*(1-t)*Y=Y \\ $$
$$ \\ Co+Io+Go+Xo-h*r=Y-(b-m)*(1-t)Y\\ $$
$$ \\Co+Io+Go+Xo-h*r=[1-(b-m)*(1-t)]Y \\ $$
$$ \\ Ye=[1/1-(b-m)(1-t)][Co+Io+Go+Xo-h*r]\\ $$





```python
Y_size = 100 

Co = 33
Io = 45
Go = 65
Xo = 3
h = 0.6
b = 0.5 # b > m
m = 0.3
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)

y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(DA_IS_K, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(L_45, color = "#404040") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas punto de equilibrio
plt.axvline(x=61,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=152, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")

# Texto agregado
    # punto de equilibrio
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(80, 180, '$DA=α0+α1*Y$', fontsize = 11, color = 'black')
plt.text(80, 170, '$α=(b-m)(1-t)$', fontsize = 11, color = 'black')
plt.text(65, 140, '$α0=(Co+Io+Go+Xo-h*r)$', fontsize = 11, color = 'black')
plt.text(70, 11, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 150, '$α_o$', fontsize = 15, color = 'black')
    # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

# Título y leyenda
ax.set(title="El ingreso de Equilibrio a Corto PlAZO", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda

plt.show()
```

Analice una Pol í t ica f iscal expansiva con una reducción de la Tasa de Tributación . El análisis debe ser intuitivo, matemático y gráfico.
- Intuitivo
$$ \\ ↓Δt→ ↑Co--> ↑DA-->DA>Y-↑Y\\ $$



- Matemáticamente


```python
Co, Io, Go, Xo, h, r, b, m, t = symbols('Co Io Go Xo h r b m t')

f = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))


df_t = diff(f, t)
df_t #∆Y/∆t
```

Considerando el diferencial de ∆t:
$$ \\∆Y/∆t=(m-b)(Co+Go+Io+Xo-hr)/(1-(1-t)(b-m)+1)^2 \\ $$


- sabiendo que b>m,entonces (m-b)<0
- Los componentes autónomos no cambian:
  $$ \\ ∆Co_=∆Io=∆Xo=∆h=∆r=0\\$$
- Cualquier número elevado al cuadrado será positivo:
    $$ \\ (1-(1-t)(b-m)+1)^2>0\\$$
- Entonces:
    $$ \\∆Y/∆t=(-)/(+)\\$$
- Dado que ∆t<0, la división de dos positivos da otro positivo:
     $$ \\ ∆Y/(-)=(-)/(+)\\ $$
     $$ \\  ∆Y=(-)(-)/(+)\\ $$
     $$ \\  ∆Y>0 \\ $$

        


- Gráficamente
