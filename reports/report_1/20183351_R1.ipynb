```python
Y_size = 100 

Co = 20
b = 0.49 

Y = np.arange(Y_size)
```

```python
def DA_A(Co, b):
    DA_A = (Co) + (b*Y)
    return DA_A

DA_A_A = DA_A(Co, b)
```

```python
y_max = np.max(DA_A_A)
fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Y, DA_A_A, color = "PINK")

ax.set(title="CONSUMO", xlabel= r'Y', ylabel= r'C')
ax.legend()

plt.show()
```
