La Tributaci√≥n Total de una econom√≠a es igual a la tasa de impuestos (t) y el ingreso agregado. La tasa de impuestos es un instrumento de la pol√≠tica fiscar que obtiene un valor mayor de 0 y menor que 1. Esto debido a que si fuese 0, el Estado no estar√≠a recibiendo nada como tributaci√≥n y si fuese 1 estar√≠a recibiendo todos los ingresos de la econ√≥mia; lo cual en ningun caso tiene sentido. Se hace uso de la f√≥rmula de tributaci√≥n para poder reemplar el valor de T en la f√≥rmula de Ingreso Disponible (ùëåùëë).

    ùëåùëë = ùëå ‚àí ùëá #Ingreso disponible

    ùëåùëë = (1 ‚àí ùë°)ùëå

Habiendo reemplazado el valor de T y hallado la nueva f√≥rmula de ùëåùëë, podemos continuar con la Funci√≥n de Consumo.

    ùêÇ = ùêÇùüé + ùêõùëåùëë #Funci√≥n de consumo 
      
El Consumo total (C) es igual al Consumo aut√≥nomo (ùêÇùüé) m√°s la multiplicaci√≥n de la Propensi√≥n Marginal a Consumir (b) y el Ingreso Disponible (ùëåùëë). La Propensi√≥n Marginal a Consumir (b) representa el cambio de consumo por unidad de cambio en el ingreso disponible, por ende b debe ser mayor que 0 y menor que 1. 

        ùëëùê∂/ùëëùëåùëë = PMgC = ùëè, donde (0 < ùëè < 1)
 
    ùêÇ = ùêÇùüé + ùêõùëåùëë
    ùêÇ = ùêÇùüé + ùêõ*(1 ‚àí ùë°)ùëå
    ùêÇ = ùêÇùüé + ùëëùê∂/ùëëùëåùëë*(1 ‚àí ùë°)ùëå #propensi√≥n marginal a consumir neta de impuestos

Habiendo reemplazado el valor de b y el valor de ùëåùëë, hemos hallado la propensi√≥n marginal a consumir neta de impuestos.

    ùêÇ = ùêÇùüé + ùêõùëåùëë
    ùêÇ/ùëåùëë = ùêÇùüé/ùëåùëë + ùêõ*ùëåùëë/ùëåùëë

Finalmente hallamos la Propensi√≥n media a consumir, la cual es igual a la Propensi√≥n Marginal a Consulmir solo cuando el Consumo Aut√≥nomo es cero. Por lo tanto, en el gr√°fico a continuaci√≥n no se tomar√° en cuenta la Propensi√≥n Media a Consumir puesto que el consumo aut√≥nomo es diferente a cero.

    ùëå = (ùê∂0 + ùêº0 + ùê∫0 + ùëã0 ‚àí ‚Ñéùëü) + (ùëè ‚àí ùëö)(1 ‚àí ùë°)ùëå

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)
```

```python
def ejer4(Co, Io, Go, Xo, h, r, b, m, t, Y):
    ejer4 = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return ejer4

ejercicio4 = ejer4(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Gr√°fico 

y_max = np.max(ejercicio4)
fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Y, ejercicio4, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #L√≠nea de 45¬∫
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")

plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$Œ±_o$', fontsize = 15, color = 'black')
plt.text(6, 4, '$45¬∞$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$‚óù$', fontsize = 30, color = '#404040')

ax.set(title="Punto de Equilibrio", xlabel= r'Y', ylabel= r'DA')
ax.legend() 

plt.show()
```
