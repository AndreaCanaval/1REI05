1. Derive y explique la función de demanda de Consumo. Replique el gráfico siguiente:
    $$ PMgC = \frac{∆ C}{∆ Y^d} = b = 0.5 $$ 

```python
## Import python libraries
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
def PMgc(Y,b):
    PMgc = 10 + (Y * b) 
    return PMgc

Y_size = 80 
bd = 0.5
Yd = np.arange(Y_size) ## Ingreso disponible
PMgc1 = PMgc(Yd,bd)

  
plt.plot(Yd, PMgc1)
plt.title('Funcion demanda de consumo ')
plt.xlabel('Yd')
plt.ylabel('PMgc')
plt.grid(True)
plt.show()
```

Derive y explique la función de demanda de Consumo. $$ PMgC = \frac{∆ C}{∆ Y^d} = b = 0.5 $$

La relacion entre ∆C y ∆Y^d es inversamente proporcional, si b tiende a 0 los valores de PMgc and ∆Y^d seran iguales

2. Derive y explique la función de demanda de Inversión. Replique el gráfico siguiente:
    $$ I = I_0 - h * r$$ 

```python
def Io(Ia,h,r):
    Io = Ia - (h * r) 
    return Io

Ia = 80  # Inversion autonoma
h = 1 # sensibilidad inversionista tasa interes fija (0,1)
r = np.arange(start=0, stop=1, step=0.01) # tasa interes real activa (0,1) 
I = Io(Ia,h,r)

plt.plot(r, I)
plt.title('Demanda de Inversion')
plt.xlabel('r')
plt.ylabel('I')
plt.grid(True)
plt.show()
```

Derive y explique la función de demanda de Inversión. $$ I = I_0 - h * r $$ 

Si la sensibilidad inversionista tasa de interes fija ($h$) tiende a cero, la inversion ($I$) y la inversion autonoma ($I_0$) seran iguales

Si la sensibilidad inversionista tasa de interes fija ($h$) tiende a uno, la inversion ($I$) y la inversion autonoma ($I_0$) tendran una diferencia aproximada de una unidad cuando $r$ tienda a uno
