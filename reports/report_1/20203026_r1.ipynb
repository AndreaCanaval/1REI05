### 4.1 Equilbrio de Y^e


Considerando que $DA=Y$ 

En la que $DA = C_o+I_o+G_o+X_o-hr+Y(b-m)(1-t)$ → $DA=𝛼^0+𝛼^1Y$

Entonces, tomando Y, la ecuasión de Ingreso de equilibrio $Y^e$:

$Y=C_o+I_o-hr+G_o+X_o+bY^d-mY^d$

$Y=\frac{1}{1-(b-m)(1-t)} (C_o+I_o+G_o+X_o-hr)$


### 4.2 Gráfico del Ingreso de Equilibrio a Corto Plazo

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)


def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))


ax.plot(Y, DA_IS_K, label = "DA", color = "purple") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #Línea de 45º


# Líneas punteadas punto de equilibrio
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")
plt.axhline(y=146, xmin= 0, xmax= 0.8, linestyle = ":", color = "grey")

# Texto agregado
    # punto de equilibrio
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')
    # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
    #Formula 𝛼^0
plt.text(72, 137, '$α_o=(C_o+I_o+G_o+X_o-hr)$', fontsize = 12, color = "grey")
    #Formula 𝛼^1
plt.text(74, 167, '$α_1=(b-m)(1-t)$', fontsize = 12, color = "grey")
    #Formula DA
plt.text(82, 193, '$DA=α_o+α_1Y$', fontsize = 11, color = "purple")
plt.text(82, 187, '$↙$', fontsize = 11, color = "purple")

# Título y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda

plt.show()
```

### 5.1 Aumento del Gasto del Gobierno en una Política Fiscal Expansiva  

#### Manera instuitiva 
 
 Sabiendo que $ΔG>0$
                 
$G↑ → DA ↑ → DA > Y → Y↑$
                     
Una política fiscal expansiva implica un incremento del ($G$). Ante esto, la demanda agregada ($DA$) a su vez sufrirá un aumento, lo que implicará un desequilibrio con el nivel de producción ($Y$). Por lo que, para voler al equilibrio se debe aumentar el nivel de producción ($Y$).   
                            

#### Manera Gráfica

```python
# Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

    # Ecuación 
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

# Definición del aumento del parametro del Gasto de Gobierno
Go = 90

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

ax.set(title = "Aumento del Gasto del Gobierno $(G_0)$", xlabel = r'Y', ylabel = r'DA')

ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #Curva sin el aumento del Gasto del Gobierno
ax.plot(Y, DA_G, label = "DA_G", color = "Purple", linestyle = 'dashed') #Curva ante el aumento del G
ax.plot(Y, L_45, color = "#404040") #línea de 45º

plt.text(0, 180, '$DA_1$', fontsize = 11.5, color = 'purple')
plt.text(0, 135, '$DA^0$', fontsize = 11.5, color = 'black') 

plt.axvline(x = 70.5, ymin= 0, ymax = 0.69, linestyle = ":", color = "grey")
plt.axhline(y = 199, xmin= 0, xmax = 0.77, linestyle = ":", color = "grey")
plt.axvline(x = 80,  ymin= 0, ymax = 0.78, linestyle = ":", color = "grey")
plt.axhline(y = 175, xmin= 0, xmax = 0.69, linestyle = ":", color = "grey")

plt.text(76, 140, '$→$', fontsize = 18, color = 'grey')
plt.text(20, 160, '$↑$', fontsize = 18, color = 'grey')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

plt.show()
```

#### Manera Matemática  

$ΔG>0$ → $ΔY$

$Y=\frac{1}{1-(b-m)(1-t)} (C_o+ΔI_o+ΔG_o+ΔX_o-hr)$

Puesto que $ΔC_o=ΔI_o=ΔX_0=Δr=0$ → $ΔG>0$

$ΔY=\frac{1}{1-(b-m)(1-t)}(ΔG)$

$ΔY=k(ΔG)$

Sabiendo que $ΔG>0$ y que $k>0$, la multiplicación de dos números positivos: 

$ΔY=(+)(+)$

$ΔY>0$

### 5.2 Reducción de la Tasa de Tributación en una Política fiscal expansiva  

#### Manera intuitiva 

Se ve un ingreso disponible ante la reducción de la tasa de impuestos, y por eso  aumenta la demanda de consumo y la demanda agregada, por lo que se aumenta el nivel de producción: 

$t↓ → Co↑ → DA ↑ → DA > Y → Y↑$



#### Manera Gráfica

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3 #tasa de tributación
r = 0.9

Y = np.arange(Y_size)

     
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)

    # Definición de la reducción de la tasa de tributación
t = 0.01

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_t = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

ax.set(title = "Reducción de la Tasa de Tributación", xlabel = r'Y', ylabel = r'DA')

ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #Curva sin la reducción de la tasa de tributación
ax.plot(Y, DA_t, label = "DA_t", color = "purple", linestyle = 'dashed') #Curva ante la reducción de la tasa de tributación
ax.plot(Y, L_45, color = "#404040") #línea de 45º


plt.axvline(x = 70.5, ymin= 0, ymax = 0.69, linestyle = ":", color = "grey")
plt.axhline(y = 176, xmin= 0, xmax = 0.7, linestyle = ":", color = "grey")
plt.axvline(x = 77,  ymin= 0, ymax = 0.75, linestyle = ":", color = "grey")
plt.axhline(y = 192, xmin= 0, xmax = 0.75, linestyle = ":", color = "grey")


plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(0, 200, '$DA_t$', fontsize = 11.5, color = 'purple')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(80, 0, '$Y_t$', fontsize = 12, color = 'purple')
plt.text(72, 45, '$→$', fontsize = 18, color = 'grey')
plt.text(20, 180, '$↑$', fontsize = 18, color = 'grey')

ax.legend()

plt.show()
```

#### Manera matemática 

```python
Co, Io, Go, Xo, h, r, b, m, t = symbols('Co Io Go Xo h r b m t')

f = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))


df_t = diff(f, t)
df_t #∆Y/∆t
```

$\frac{ΔG}{Δt}=\frac{(m-b)(Co+Go+Io+Xo-hr)}{(1-(1-t)(b-m)+1)^2}$

$ΔC_o=ΔI_o=ΔX_o=Δh=Δr=0$

$(1-(1-t)(b-m)+1)^2>0$

$\frac{ΔY}{Δt}=\frac{(-)}{(+)}$

Como $Δt<0$

$\frac{ΔY}{(-)}=\frac{(-)}{(+)}$

$ΔY=\frac{(-)(-)}{+}$

$ΔY>0$
