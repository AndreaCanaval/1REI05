### CUARTA PREGUNTA

La fórmula de la demanda agregada tiene la siguiente forma: 

$$ DA = C + I + G + X - M $$

Y se compone de:

Función de consumo:$$C = C_o + bY^d$$ 
Función Inversión: $$I = I_o − hr$$
Gasto del Gobierno: $$ G = G_o $$
Tributación: $$ T = tY $$
Exportaciones: $$X = X_o$$
Importaciones: $$M = mY^d$$

Y, por ello, también puede ser representada de la siguiente forma: 

$$DA = Co + bY^d + Io - hr + Go + tY + Xo - mY^d$$

Factorizando: 

$$DA = Co + Io + Go + Xo - hr + bY^d + tY- mY^d$$

$$DA = Co + Io + Go + Xo - hr + (b-m)(1-t)Y$$

Cuando nos encontramos en equilibrio, $Y = DA$. Entonces la ecuación se plantea: 

$$Y = Co + Io + Go + Xo - hr + (b-m)(1-t)Y$$

$$Y - (b-m)(1-t)Y= Co + Io + Go + Xo - hr$$

$$[1 - (b-m)(1-t)]Y= Co + Io + Go + Xo - hr$$

$$ Y= \frac {1} {1 - (b - m)(1 - t)} (Co + Io + Go + Xo - hr)$$

Gráfico del ingreso de equilibrio:

```python
# Parámetros

Y_size = 100 

Co = 20
Io = 35
Go = 70
Xo = 3
h = 0.6
b = 0.7 # b > m
m = 0.2
t = 0.5
r = 0.8

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Recta de 45°

a = 2.5

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#FF5733") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas punto de equilibrio
plt.axvline(x=56.675555555555555,  ymin= 0, ymax= 0.57, linestyle = ":", color = "grey")
plt.axhline(y=141.68888888888887, xmin= 0, xmax= 0.57, linestyle = ":", color = "grey")

# Texto agregado
    # punto de equilibrio
plt.text(0, 149, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(58, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 116, '$α_o$', fontsize = 15, color = 'black')
plt.text(80, 160, '$DA = α_o + α_1Y$', fontsize = 15, color = 'black')
plt.text(75, 129, '$α_1 = (b-m)(1-t)$', fontsize = 13, color = 'black')
plt.text(65, 60, '$α_o = (C_o + I_o + G_o + X_o - hr$', fontsize = 13, color = 'black')

    # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

# Título y leyenda
ax.set(title="Ingreso de equilibrio a corto plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda

plt.show()
```

```python
# líneas punteadas autómaticas

    # definir la función line_intersection
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y

    # coordenadas de las curvas (x,y)
A = [DA_IS_K[0], Y[0]] # DA, coordenada inicio
B = [DA_IS_K[-1], Y[-1]] # DA, coordenada fin

C = [L_45[0], Y[0]] # L_45, coordenada inicio
D = [L_45[-1], Y[-1]] # L_45, coordenada fin

    # creación de intersección

intersec = line_intersection((A, B), (C, D))
intersec
```

1. Política Fiscal Expansiva con aumento del Gasto del Gobierno 

**Intuición:** 
$$↑G_o → ↑DA → DA > Y → ↑Y$$

**Matemática:**

Si se sabe que $ΔG_o > 0$, entonces el $ΔY$:

$$ Y= \frac {1} {1 - (b - m)(1 - t)} (Co + Io + Go + Xo - hr)$$

Se reemplaza con el multiplicador keynesiano: 

$$ Y= k(Co + Io + Go + Xo - hr)$$

Y reemplazamos con los diferenciales:

$$ ΔY= k(ΔCo + ΔIo + ΔGo + ΔXo - Δhr)$$

Sin embargo, como el algunos de los coeficientes no se han dado cambios, estos no son considerados: 

$$ ΔY= k(ΔGo)$$

Recordando que $ΔGo > 0$ y que $k > 0$, la operación de signos sería: 

$$ ΔY=(+)(+)$$

$$ ΔY>0$$

**Gráfico:**

**2. Política fiscal expansiva con una reducción de la Tasa de Tributación**

**Intuición:**

$$ t↓→C_o↑→DA↑→DA>Y→ Y↑$$

$$ t↓→M↑→DA↓→DA<Y→ Y↓ $$

**Matemática:**
Si se sabe que $Δt < 0$, entonces el $ΔY$:

$$ Y= \frac {1} {1 - (b - m)(1 - t)}(Co + Io + Go + Xo - hr)$$

El diferencial de $∆t$:

```python
Co, Io, Go, Xo, h, r, b, m, t = symbols('Co Io Go Xo h r b m t')

f = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))


df_t = diff(f, t)
df_t #∆Y/∆t
```

$$ \frac{∆Y} {∆t} = \frac {(−𝑏+𝑚)(𝐶𝑜+𝐺𝑜+𝐼𝑜+𝑋𝑜−ℎ𝑟)}{(−(1−𝑡)(𝑏−𝑚)+1)^2} $$

Teniendo en cuenta que:
$$(m-b)<0$$ porque $b>m$

Que los diferenciales de algunos componentes son iguales a cero porque no varian:
$$∆𝐶𝑜 = ∆𝐺𝑜 = ∆𝐼𝑜 = ∆𝑋𝑜 = ∆ℎ = ∆𝑟 = 0$$

Y que todo número elevado al cuadrado es positivo: 
$$ (−(1−𝑡)(𝑏−𝑚)+1)^2 > 0 $$

Por ley de signos podemos determinar: 

$$ \frac{∆Y} {∆t} = \frac {(-)}{(+)} $$

Y dado que $Δt < 0$, se puede reemplazar:

$$ \frac{∆Y} {(-)} = \frac {(-)}{(+)} $$

$$ ∆Y= \frac {(-)(-)}{(+)} $$

$$ ∆Y > 0 $$


**Gráfico:**
