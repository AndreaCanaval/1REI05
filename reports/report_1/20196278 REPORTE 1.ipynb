- En primer lugar, definimos (DA) es entendida como la demanda agregada que es necesario cubrir para encontrar el punto de equilibrio con la oferta. Esta demanda está dada por los elementos del consumo (C) más inversión (I) más gasto del gobierno (G) más tributación t(Y) más las exportaciones netas (X-M)
$$DA=C+I+G+t(Y)+(X-M)$$
- Acon tinuación reemplazamos cada función
$$DA= Co+b(Yd)+Io-h(r)+Go+Xo-M(Yd)$$
- Factorizando y reemplazando Yd=(1-t)Y
$$DA=Co+Io-h(r)+Go+Xo + (b-m)(1-t)Y$$
- El punto de equilibrio estaría dado duando DA sea igual a Ye
$$DA=Ye$$
- Se factoriza y se identifican la función Ye, su intercepto y su pendiente
$$pendeinte (α1)= (b-m)(1-t)$$
$$intercepto(α0)= Co+Io-h(r)+Go+Xo$$
- Finalmente
    $$Ye=1/(1-(b-m)(1-t))*(Co+Io+Go+Xo-hr)$$

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
# Dimensiones del gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #Línea de 45º

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Líneas punteadas punto de equilibrio
plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")

# Texto agregado
    # punto de equilibrio
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')

plt.text(90, 230, '↓', fontsize=20, color='black')
plt.text(75, 240, 'Da= α0 + α1 (Y)', fontsize=20, color='black')
plt.text(75, 200, 'α1=(b-m)(1-t)', fontsize=20, color='black')
plt.text(75, 80, 'α0=(Co+Io+Go+Xo-hr)', fontsize=20, color='black')
plt.text(75, 100, '↑', fontsize=20, color='black')
    # línea 45º
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

# Título y leyenda
ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend() #mostrar leyenda

plt.show()
```
