“The lean times in the Peruvian economy” o “Las vacas flacas en la economía peruana” es un artículo publicado en 2017 por el economista peruano Óscar Dancourt. El artículo pretende responder qué tanto ha impactado de manera adversa el choque externo de la caída de los precios de las materias primas exportadas por el Perú, durante los años 2014 y 2015, y evaluar si las políticas monetarias y fiscales adoptadas por el Banco Central de Reserva y el Ministerio de Economía y Finanzas, respectivamente, son las correctas para afrontar las “épocas de vacas flacas” presentes a lo largo de la historia macroeconómica del Perú. 
Para ello, el autor analiza el impacto del choque externo adverso en diversos ámbitos, como en la caída de la producción y la inversión en el Perú, la reducción de los créditos al sector privado otorgados en moneda extranjera, el alza del tipo de cambio y el consiguiente proceso inflacionario. En adición, Dancourt evalúa y critica la política monetaria adoptada por el Banco Central de Reserva, que consistió en vender gran parte de las reservas de divisas, sin recurrir a la reducción de la tasa de interés de referencia. Asimismo, realiza lo propio con la política fiscal del Ministerio de Economía y Finanzas. Esta se basó en rebajar los estándares laborales y medioambientales para supuestamente estimular la inversión privada, así como en reducir la inversión pública e implementar rebajas impositivas. 
Este enfoque de investigación resulta útil en tanto permite conocer a detalle los procesos macroeconómicos que ocurren en cadena como resultado de la caída de los precios de las materias primas exportadas por el Perú. La principal fortaleza del enfoque utilizado por Dancourt es que explica detalladamente las consecuencias del choque externo en diversos indicadores macroeconómicos. Su enfoque se orienta hacia un modelo keynesiano. Utiliza el concepto de la demanda agregada para demostrar la caída de la inversión y el gasto de gobierno como consecuencia de la reducción del ingreso disponible. Asimismo, Dancourt ilustra el impacto inflacionario del alza del tipo de cambio, considerando las particularidades de la economía peruana, la cual contiene bienes y servicios cuyos precios dependen de la fluctuación del precio del dólar. No obstante, el documento no presenta un contraste de ideas entre otras teorías macroeconómicas, lo cual constituye una debilidad en el enfoque utilizado.
El autor contribuye positivamente a responder la pregunta de investigación, pues no sólo realiza un análisis de las causas y consecuencias del choque externo en el crecimiento de la economía peruana, sino que también evalúa los aciertos y desaciertos de la política monetaria y fiscal y propone alternativas basadas en una perspectiva keynesiana. Dancourt sostiene que el BCR debió optar por reducir la tasa de interés referencial en vez de vender gran cantidad de las reservas en moneda extranjera y la rebaja de la tasa de encaje intentando estimular el crédito privado. Con respeto a la política fiscal, el autor contribuye sugiriendo un aumento de la inversión pública, alegando que el Estado peruano puede recurrir al Fondo de Estabilización Fiscal o el endeudamiento debido a su posición privilegiada en el sistema financiero internacional. Así, Dancourt avanza en el conocimiento sobre esta reciente recesión económica.
El artículo, en sus conclusiones, plantea una serie de retos para el gobierno de Pedro Pablo Kuczynski, el cual no cumplió con afrontarlos. En ese sentido, un siguiente paso para profundizar en la pregunta de investigación sería no tanto centrarse en las causas y consecuencias de la recesión, sino en idear una política fiscal y monetaria que pueda rescatar al Perú de la desaceleración y posterior crisis económica en la cual se encuentra, no sólo desde un enfoque keynesiano, sino incorporando otros puntos de vista al debate.

1. Función de demanda de Consumo
La función demanda de consumo se obtiene calculando la suma del Consumo Autónomo, es decir, los consumos fijos, y otro consumo adicional, determinado por la propensión marginal a consumir, que va en función al ingreso disponible.

```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
from sympy import *
import pandas as pd
from causalgraphicalmodels import CausalGraphicalModel
```

```python
Yd_size = 100 

Co = 35 
b = 0.8 

Yd = np.arange(Yd_size)


def C(Co, b, Yd):
    C = (Co+(b*Yd))
    return C

C1 = C(Co, b, Yd)
C1
```

```python
Yd_max = np.max(C1)
fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Yd, C1, label = "C", color = "green") 

plt.text(45, 100, '$PMgC= △C/△Yᵈ = b = 0.5 $', fontsize = 12, color = 'black')



ax.set(title = "Función de demanda de Consumo", xlabel = 'Yᵈ', ylabel = 'C')
ax.legend()

plt.show()
```

2. Función de inversión
La función de inversión es determinada por la suma de la inversión autónoma, es decir, aquella que es fija, con la inversión dependiente de la sensibilidad de los inversionistas antes las variaciones en la tasa de interés de referencia.

Io = I + hr

```python
r_size = 100 

Io = 40
h = 0.7

r = np.arange(r_size)

def I(Io, r, h):
    I = Io-h*r
    return I

I1= I(Io, r, h)
I1
```

```python
r_max = np.max(I1)
fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Yd, I1, label = "I", color = "#3D59AB") 

plt.text(70, -1, '$I= Io - hr$', fontsize = 12, color = 'black')
plt.text(68, -4, '$↓$', fontsize = 18, color = 'green')


ax.set(title = "Función de demanda de Inversión", xlabel = 'r', ylabel = 'I')
ax.legend()

plt.show()
```

4. Encontrar el nivel de Ingreso de equilibrio a partir de las ecuaciones que componen la demanda agregada:
                                𝐷𝐴 = 𝐶o + 𝑏𝑌ᵈ + 𝐼o − ℎ𝑟 + 𝐺o + 𝑋o − 𝑚𝑌ᵈ
                                𝐷𝐴 = (𝑏 − 𝑚)(1 − 𝑡)𝑌 + 𝐶o + 𝐼o − ℎ𝑟 + 𝐺o + 𝑋o
                                𝐷𝐴 = (𝐶o + 𝐼o + 𝐺o + 𝑋o − ℎ𝑟) + (𝑏 − 𝑚)(1 − 𝑡)𝑌
                                𝐷𝐴 = 𝛼o + 𝛼1𝑌

En esta ecuación, el intercepto es α0 = (Co + Io + Go + Xo − hr) y la pendiente es α1 = (b − m)(1 − t)

Si en equilibrio, el ingreso agregado (Y) es igual a la demanda agregada (DA), la ecuación del ingreso de equilibrio a corto plazo será la siguiente:

                                 Y = DA

                                 𝑌 = (𝐶o + 𝐼o + 𝐺o + 𝑋o − ℎ𝑟) + (𝑏 − 𝑚)(1 − 𝑡)𝑌

                                 [1 − (𝑏 − 𝑚)(1 − 𝑡)]𝑌 = 𝐶o + 𝐼o + 𝐺o + 𝑋o − ℎ𝑟

                                 𝑌ᵉ =          1            
                                     ——————————————————— (𝐶o + 𝐼o + 𝐺o + 𝑋o − ℎ𝑟)
                                      1 − (𝑏 − 𝑚)(1 − 𝑡)

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)


def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

```python
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(12, 8))


ax.plot(Y, DA_IS_K, label = "DA", color = "orange") 
ax.plot(Y, L_45, color = "purple") 


ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())


plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=146, xmin= 0, xmax= 1, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")


plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')
plt.text(88, 176, '$α_₁= (b-m)(1-t)$', fontsize = 11, color = 'black') 
plt.text(88, 190, '$↙$', fontsize = 18, color = 'red')
plt.text(90, 200, '$DA= αo+α₁ Y $', fontsize = 11.5, color = 'black') 
plt.text(75, 137, '$↖$', fontsize = 18, color = 'red')
plt.text(75, 130, '$αo= (Co + Io + Go + Xo - hr)$', fontsize = 11.5, color = 'black') 
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')


ax.set(title="Ingreso de equilibrio a corto plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend()

plt.show()
```
