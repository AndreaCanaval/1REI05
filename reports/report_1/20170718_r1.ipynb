Ejercicio1: Derive y explique la funci√≥n de demanda de Consumo

Para poder calcular la funci√≥n Consumo (C), debemos tomar en cuenta que la Tributaci√≥n total (T); es decir todo lo que se recauda en el pa√≠s, es igual a la tasa impositiva (t) por el Ingreso Agregado (Y) de la econom√≠a:


$$ T = tY $$


En este contexto, suponiendo que todas las transferencias (TR) son iguales a cero, el ingreso neto de la tributaci√≥n es el Ingreso disponible (ùëå·µà):


$$ Y^d = Y - T $$

$$ Y^d = (1 - t) Y $$


Una parte de este Ingreso disponible (ùëå·µà) se utiliza para los gastos de consumo, por lo que resulta la siguiente f√≥rmula:


$$ C = C_0 + bY^d $$


De esta manera, podemos comprender que la funci√≥n de consumo depende no solo de una parte de este Ingreso Disponible (ùëå·µà), sino tambi√©n del Consumo aut√≥nomo (Co) y la propensi√≥n marginal a consumir (b) que representa el cambio en el consumo por unidad de cambio en el ingreso disponible; la cual es mayor que cero y menor que uno, 0 < b < 1.
    

```python
Yd_size = 100 

Co = 10
b = 0.5 

Yd = np.arange(Yd_size)
```

```python
def C(Co, b, Yd):
    C= Co + b*Yd
    return C
C= C(Co, b, Yd)
```

```python
# Gr√°fico

# Dimensiones del gr√°fico
y_max = np.max(C)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(Yd, C, label = "C", color = "#3D59AB")

# L√≠neas punteadas punto de equilibrio
plt.axvline(x=40,  ymin= 0, ymax= 0.60, linestyle = ":", color = "grey")
plt.axhline(y=40, xmin= 0, xmax= 0.41, linestyle = ":", color = "grey")

plt.axvline(x=30,  ymin= 0, ymax= 0.60, linestyle = ":", color = "grey")
plt.axhline(y=30, xmin= 0, xmax= 0.41, linestyle = ":", color = "grey")

plt.axvline(x=20,  ymin= 0, ymax= 0.60, linestyle = ":", color = "grey")
plt.axhline(y=20, xmin= 0, xmax= 0.41, linestyle = ":", color = "grey")

plt.axvline(x=10,  ymin= 0, ymax= 0.60, linestyle = ":", color = "grey")
plt.axhline(y=10, xmin= 0, xmax= 0.41, linestyle = ":", color = "grey")

# T√≠tulo y leyenda
ax.set(title="Funci√≥n de demanda de Consumo", xlabel= r'Yd', ylabel= r'C')
ax.legend() 
```

Ejercicio2: Derive y explique la funci√≥n de Inversi√≥n.

Para poder calcular la funci√≥n gasto de Inversi√≥n (I), debemos tomar en cuenta que esta depende inversamente de la tasa de inter√©s real (r); y donde el comonente aut√≥nomo (Io), que vendr√≠a a ser la inversi√≥n fija, depende de las expectativas de los inversionistas:


$$ I = I_0 - hr$$


En esta ecuaci√≥n, el "h" vendr√≠a a ser la sensibilidad de los inversionistas ante la subida o reducci√≥n de la tasa de inter√©s real "r". Adem√°s, ambas se encuentran en un rango mayor a cero y menor que 1; 0 < h <1 y 0 < r <1.


```python
# Par√°metros

r_size = 100 


Io = 40
h = 0.5
r = 0.8

r = np.arange(r_size)
```

```python
def I(Io, h, r):
    I= Io - h*r
    return I
I= I(Io, h, r)
```

```python
# Gr√°fico

# Dimensiones del gr√°fico
y_max = np.max(I)
fig, ax = plt.subplots(figsize=(10, 8))

# Curvas a graficar
ax.plot(r, I, label = "I", color = "#3D59AB")

# T√≠tulo y leyenda
ax.set(title="Funci√≥n de demanda de Inversi√≥n", xlabel= r'r', ylabel= r'I')
ax.legend() 

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto agregado
plt.text(80, 7, '$I=I_o-hr$', fontsize = 15, color = 'black')
```

A partir de las siguientes ecuaciones:
    
    
$$ C = C_0 + bY^d $$

$$ I = I_0 - hr $$

$$ G = G_0 $$

$$ T = tY $$

$$ X = X_0 $$

$$ M = mY^d $$

$$ DA = C + I + G + X - M $$


Encuentre el nivel de Ingreso de equilibrio $Y^e$

La ecuaci√≥n de equilibrio para el Ingreso Agregado se deriva de la condici√≥n de equilibrio donde el ingreso es igual a la demanda agregada: $DA = Y$:


$$ DA = C + I + G + X - M $$


Con los datos previos, reemplazos los datos anteriores en la funci√≥n de la Demanda Agregada (DA):


$$ Y = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $$

$$ DA = C_0 + I_0 + G_0 + X_0 - hr + Y(b - m)(1 - t) $$


En donde el Intercepto (Œ±0) y la pendiente (Œ±1) son iguales a:


$ Œ±_0 = (C_0 + I_0 + G_0 + X_0 -hr)$

$ Œ±_1 = (b - m)(1 - t) $


De forma corta:
    
$$ DA = Œ±_0 + Œ±_1Y $$


Considerando la condici√≥n de equilibrio $Y = DA$, la ecuaci√≥n del ingreso de equilibrio a corto plazo es:


$$ Y = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $$

$$ Y = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

En este sentido el multiplicador Keynesiano (K) es igual a:

$$ K = \frac{1}{1 - (b - m)(1 - t)}$$

Otra forma de verlo:


$$ Y = K (C_0 + I_0 + G_0 + X_0 - hr) $$




```python
# Par√°metros

Y_size = 100 

Co = 30
Io = 45
Go = 65
Xo = 2
h = 0.6
b = 0.7 
m = 0.3
t = 0.4
r = 0.8

Y = np.arange(Y_size)

# Ecuaci√≥n de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Recta de 45¬∞

a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)
```

$$ ‚àÜY = (+)(+) $$

Analice una Pol√≠tica fiscal expansiva con una reducci√≥n de la Tasa de Tributaci√≥n. El an√°lisis debe ser intuitivo, matem√°tico y gr√°fico.

#### - Pol√≠tica fiscal expansiva con una reducci√≥n de la Tasa de Tributaci√≥n $(t)$:

- Intuici√≥n: ¬øcontradicci√≥n?

$$ t‚Üì ‚Üí Co‚Üë ‚Üí DA‚Üë ‚Üí DA > Y ‚Üí Y‚Üë $$
$$ t‚Üì ‚Üí M‚Üë ‚Üí DA‚Üì ‚Üí DA < Y ‚Üí Y‚Üì $$

- Matem√°ticamente: $‚àÜt < 0  ‚Üí  ¬ø‚àÜY?$

Considernado el diferencial de $‚àÜt$:

$$ \frac{‚àÜY}{‚àÜt} = \frac{(m-b)(Co + Go + Io + Xo - hr)}{(1-(1-t)(b-m)+1)^2} $$

- Sabiendo que b > m, entonces $(m-b) < 0$
- Los componentes aut√≥nomos no cambian: $‚àÜC_0 = ‚àÜI_0 = ‚àÜX_0 = ‚àÜh = ‚àÜr = 0$
- Cualquier n√∫mero elevado al cuadrado ser√° positivo: $ (1-(1-t)(b-m)+1)^2 > 0 $

Entonces:

$$ \frac{‚àÜY}{‚àÜt} = \frac{(-)}{(+)} $$

Dado que $‚àÜt < 0$, la divisi√≥n de dos positivos da otro positivo:

$$ \frac{‚àÜY}{(-)} = \frac{(-)}{(+)} $$

$$ ‚àÜY = \frac{(-)(-)}{(+)} $$

$$ ‚àÜY > 0 $$
